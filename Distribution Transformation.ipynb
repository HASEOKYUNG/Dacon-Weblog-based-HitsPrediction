## Import

```python
import numpy as np
np.random.seed(2024)
from scipy.stats import skew
from scipy.stats import beta
from scipy.stats import boxcox
import matplotlib.pyplot as plt
import seaborn as sns
```

## $\blacktriangleright$ Log Transformation
- `양수`인 데이터이며 지수분포, 카이제곱분포, 감마분포 등 `왼쪽`으로 치우친 분포에 사용한다.
- 지나치게 큰 값이 있는 경우에 유용하다.

```python
# 지수분포(Left-Skewed)의 log 변환
data = np.random.exponential(scale=2, size=1000)
transformed = np.log(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Exponential Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Log-Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```

## $\blacktriangleright$ Square root Transformation
- 지수분포, 카이제곱분포, 감마분포 등 왼쪽으로 치우친 경우에 사용한다.
- 포아송 분포처럼 분산이 평균에 비해 큰 경우에 유용하다.

```python
# 카이제곱 분포의 제곱근 변환
data = np.random.chisquare(3, size=1000)
transformed = np.sqrt(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Chisquare Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Square Root-Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```

## $\blacktriangleright$ Box-Cox Transformation
- 데이터가 양수이고 0이 없는 경우에 사용할 수 있다.
- 베타분포 같이 `오른쪽`으로 치우친 경우에도 사용할 수 있다.
- 정규성 가정을 만족시킬 때 가장 많이 적용하는 기법이다.

```python
# 감마분포(Left-Skewed)의 box-cox 변환
data = np.random.gamma(shape=2, scale=2, size=1000)
transformed, _ = boxcox(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Gamma Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Box-Cox Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```

```python
# 베타분포(Right-Skewed)의 log 변환
data = beta.rvs(18, 2, size=1000)
transformed, _ = boxcox(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Beta Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Box-Cox Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```

## ※ Reciprocal Transformation & Exponential Transformation
- 값들간 차이를 주기 위해서 비대칭으로 변환하는 방법이다.
- 큰 값들간 차이를 벌리고자 하면 -1을 곱하면 된다.

```python
# 균일분포의 역수 변환
data = np.random.uniform(low=1, high=5, size=1000)
transformed = 1 / data

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Uniform Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Reciprocal-Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```

```python
# 균일분포의 지수변환
transformed = -1*np.exp(data)

fig, ax = plt.subplots(1, 2, figsize=(10, 4), sharey=True)
sns.histplot(data, bins=30, color='olive', alpha=0.7, ax=ax[0])
ax[0].set_title(f"Uniform Distribution\n(skew: {skew(data)})")
sns.histplot(transformed, bins=30, color='dimgray', alpha=0.7, ax=ax[1])
ax[1].set_title(f'Exponential-Transformed Data\n(skew: {skew(transformed)})')
plt.show()
```
